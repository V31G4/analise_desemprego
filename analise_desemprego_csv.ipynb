```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

```python
df = pd.read_csv("C:/Users/Camil/Downloads/desemprego_estados.csv")
df.head()
```

```python
print("Estados:", df['Estado'].unique())
print("Anos:", df['Ano'].unique())
```

```python
plt.figure(figsize=(10, 6))
for estado in df['Estado'].unique():
    dados_estado = df[df['Estado'] == estado]
    plt.plot(dados_estado['Ano'], dados_estado['Taxa (%)'], label=estado)

plt.title("Taxa de Desemprego por Estado (exemplo)")
plt.xlabel("Ano")
plt.ylabel("Taxa de Desemprego (%)")
plt.legend()
plt.tight_layout()
plt.show()
```
